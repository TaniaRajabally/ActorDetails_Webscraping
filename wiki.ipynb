```
import operator
from functools import reduce
# img_class = soup.find_all('div', attrs={"class":"lister-item-image"})
# img=img_class.find('img')
# img
# urlss= soup.select(".div-col columns column-width")
# print(urlss)
name=soup.find_all('div', attrs={"class":"div-col columns column-width"})
#print(name)
# for ultag in soup.find_all('ul', ):
#      for litag in ultag.find_all('li'):
#              print (litag.text)
ur=[n.find('ul') for n in name]
#print(ur)
u=[n.find_all('li') for n in ur]
#print(u)
a=[]
for i in u:
  url = [n.find('a') for n in i]
  urls=[im.get('href') for im in url]
  a.append(urls)
flattened_list = [y for x in a for y in x]
flattened_list
# print(urls)
#urls
#img_url = img_class[0].get('src')
#img_url
```

```
from bs4 import BeautifulSoup
actors=[]
images=[]
for i in flattened_list:
  #print(i)
  page = requests.get("https://en.wikipedia.org/"+i)
  #print(page.content)
  soup = BeautifulSoup(page.content, 'html.parser')
  img=soup.find_all('a', attrs={"class":"image"})
  #print(img)
  imgs=["https://en.wikipedia.org/"+im.get('href') for im in img]
  if 'https://en.wikipedia.org//wiki/File:Question_book-new.svg' in imgs:
    imgs.remove('https://en.wikipedia.org//wiki/File:Question_book-new.svg')
  images.append(imgs)
  name=soup.find_all('h1', attrs={"id":"firstHeading"})
  names = [n.get_text() for n in name]
  actors.append(names)
images
actors
```

```
images
```

```
from flashtext import KeywordProcessor
pers=[]
for i in flattened_list:
  #print(i)
  page = requests.get("https://en.wikipedia.org/"+i)
  #print(page.content)
  soup = BeautifulSoup(page.content, 'html.parser')
  text=soup.find_all('p')
  #text
  texts = [n.get_text() for n in text]
  #print(texts)
  keyword_processor = KeywordProcessor()
  # keyword_processor.add_keyword(<unclean name>, <standardised name>)
  k=['beautiful','good looks','yoga','beauty','radio jockey','host','hosting','loner','memorable','married','philanthropist','comedy','successful','respected','comedic','director','writer','painter','photographer','screenwriter','dedicated','modelling','tall','sensitive','beautiful','intelligent','creative','outstanding','gorgeous','known for her work','known for his work','powerful','powerful performances','actor','actress','producer','stand-up comedian','beauty with brains','TV anchor','singer','writer','humble','brave','courageous','stubborn','loyal','guile','selfish','generous','respectful','imaginative','independent','honest','daring','lazy','responsible','cautious','polite','bossy','general']
  hi=[]
  for i in k:
    keyword_processor.add_keyword(i)
  #keyword_processor.add_keyword('Bay Area')
  for i in texts:
    #print(i)
    keywords_found = keyword_processor.extract_keywords(i) 
    keywords_found = list(dict.fromkeys(keywords_found))
    if keywords_found==False:
      hi.append('')
    else:
      hi.append(keywords_found)
  #hi
  traits = [y for x in hi for y in x]
  traits
  pers.append(traits)
pers
# ['New York', 'Bay Area']
```
