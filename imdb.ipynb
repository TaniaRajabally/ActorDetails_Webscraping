```
from bs4 import BeautifulSoup
soup = BeautifulSoup(page.content, 'html.parser')
```

```
#ans=soup.find(class_="lister-item mode-detail")
#print(ans)
```

```
name= soup.select(".lister-item-header")
names = [n.get_text() for n in name]
#names
#names[1]
names[:] = [s.replace('\n', '') for s in names]
names
```

```
# img_class = soup.find_all('div', attrs={"class":"lister-item-image"})
# img=img_class.find('img')
# img
images= soup.select(".lister-item-image")
img = [n.find('img') for n in images]
imgs=[im.get('src') for im in img]
imgs
#img_url = img_class[0].get('src')
#img_url
```

```
cont= soup.select(".lister-item-content")
content = [n.find('p',class_=False) for n in cont]
#content
contents=[im.get_text() for im in content]
#contents
contents[:] = [s.replace('\n', '') for s in contents]
contents
```

```
!pip install flashtext
```

```
from flashtext import KeywordProcessor
keyword_processor = KeywordProcessor()
# keyword_processor.add_keyword(<unclean name>, <standardised name>)
k=['beautiful','good looks','yoga','beauty','radio jockey','host','hosting','loner','memorable','married','philanthropist','comedy','successful','respected','comedic','director','writer','painter','photographer','screenwriter','dedicated','modelling','tall','sensitive','beautiful','intelligent','creative','outstanding','gorgeous','known for her work','known for his work','powerful','powerful performances','actor','actress','producer','stand-up comedian','beauty with brains','TV anchor','singer','writer','humble','brave','courageous','stubborn','loyal','guile','selfish','generous','respectful','imaginative','independent','honest','daring','lazy','responsible','cautious','polite','bossy','general']
hi=[]
for i in k:
  keyword_processor.add_keyword(i)
#keyword_processor.add_keyword('Bay Area')
for i in contents:
  #print(i)
  keywords_found = keyword_processor.extract_keywords(i) 
  keywords_found = list(dict.fromkeys(keywords_found))
  if keywords_found==False:
    hi.append('')
  else:
    hi.append(keywords_found)
hi
# ['New York', 'Bay Area']
```

```
import pandas as pd
actor = pd.DataFrame({
    "Name": names,
    "Image": imgs,
    "Description": contents,
    "Personality Traits":hi,
})
actor
```

```
import csv
actor.to_csv('Actor.csv')
```
